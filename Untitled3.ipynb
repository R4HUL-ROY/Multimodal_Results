```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import chain
```

```python
xl = pd.ExcelFile("/content/Final_result_excel.xlsx")
```

```python
pre = xl.parse("precision", index_col=None)
rec = xl.parse("recall", index_col=None)
f1 = xl.parse("f1", index_col=None)
acc = xl.parse("accuracy",index_col=None)
```

```python
sns.set(font_scale = 2)
plt.figure(figsize=(30,20))
```

# Datasets

```python
acc
```

```python
pr=[]
a=[]
for i in range(11):
    for j in range(10):
        a.append(acc[acc.columns[i]].iloc[j])
val = [list(a)]
#val = [list(acc.to_numpy().flatten())]
val = list(chain.from_iterable(val))
pr.append(val)
dataX = list(chain.from_iterable(pr))
```

```python
ds = acc.columns  #dataset names
dsName = []
for i in range(11):
  for j in range(10):
    dsName.append(ds[i])

from itertools import chain 
dataName = list(chain.from_iterable([dsName]))
```

```python
steps = [69,53,90,29,12,47,79,8,43,82]
st = []
for i in range(11):
  for j in range(10):
    st.append(steps[j])

from itertools import chain 
stepName = list(chain.from_iterable([st]))
```

```python
df2 = pd.DataFrame({
    "data": dataX,
    "Dataset":dataName,
    "step":stepName
})
```

```python
ax = sns.violinplot(x="Dataset", y="data", data=df2, palette="muted")


ax.xaxis.set_label_text("Dataset",fontsize=20)
ax.yaxis.set_label_text("Accuracy",fontsize=20)

ax.set_xticklabels(labels=acc.columns)

from google.colab import files
plt.savefig('accuracy.eps', dpi = 1200, format = 'eps')
# files.download("accuracy.eps")
```

# accuracy, precision, recall , f1 score  (glove and VGG16)

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
cols = ['image(VGG16)', 'Text(GloVe)', 'Multimodal(glove+vgg16)']
```

```python
#dataset names
dsName = []
for i in range(3):
  for j in range(10):
    dsName.append(cols[i])

from itertools import chain 
dataName = list(chain.from_iterable([dsName+dsName+dsName+dsName]))
```

```python
#steps in X axis = number of metric
s1=[]
s2=[]
s3=[]
s4=[]

for t in range(30):
  s1.append("Precision")
  s2.append("Recall")
  s3.append("F1-score")
  s4.append("Accuracy")
xstep = []
xstep = list(chain.from_iterable([s1+s2+s3+s4]))
```

```python
# number of data values

pr = []
a=[]
b=[]
c=[]
d=[]
for i in range(3):
    for j in range(10):
        a.append(pre[cols[i]].iloc[j])
for i in range(3):
    for j in range(10):
        b.append(rec[cols[i]].iloc[j])
for i in range(3):
    for j in range(10):
        c.append(f1[cols[i]].iloc[j])
for i in range(3):
    for j in range(10):
        d.append(acc[cols[i]].iloc[j])

val = [list(a) + list(b) + list(c) + list(d)]
val = list(chain.from_iterable(val))
pr.append(val)
dataX = list(chain.from_iterable(pr))
```

```python
dfn = pd.DataFrame({
    "data": dataX,
    "Dataset": dataName,
    "step": xstep
})
```

```python
sns.set(font_scale = 2)
plt.figure(figsize=(30,20))

ax = sns.boxplot(x="step", y="data", hue="Dataset",

                 data=dfn,  dodge=True)
#ax.axhline(0, ls='--')

ax.xaxis.set_label_text("Metric",fontsize=20)
ax.yaxis.set_label_text("Performance",fontsize=20)

ax.set_xticklabels(labels=["Precision","Recall","F1-score", "Accuracy"], rotation='horizontal')
#plt.legend(loc = 'upper right', title = 'Dataset', fontsize='12', title_fontsize='15')
plt.legend(loc = 'upper center', title = 'Models', fontsize='25', title_fontsize='30', bbox_to_anchor = (0.5, 1.125), fancybox=True, shadow=True, ncol=6)

plt.savefig('PrecisionRecallF1Score.eps', dpi = 1200, format = 'eps')
# files.download("PrecisionRecallF1Score.eps")
```
